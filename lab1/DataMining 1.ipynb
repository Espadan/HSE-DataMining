Разделим выборку на test и train:

Применим k-nn модель для классификации:

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

С помощью confusion_matrix можно увидеть сколько ответов модель дала правильно, а сколько с ошибкой превого/второго рода.

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_predict))
```

Precision и recall используются для минимизации ошибки первого/второго рода. F1 общая метрица, для минимизации всех ошибок.

```python
from sklearn.metrics import roc_curve

sns.set(font_scale=1.5)
sns.set_color_codes("muted")

plt.figure(figsize=(10, 8))
fpr, tpr, thresholds = roc_curve(y_test, knn.predict_proba(X_test)[:,1], pos_label=1)
lw = 2
plt.plot(fpr, tpr, lw=lw, label='ROC curve ')
plt.plot([0, 1], [0, 1])
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve')
plt.savefig("ROC.png")
plt.show()
```

ROC-curve подходит для бинарных классификаторов, хорошо подходит для неравномерной выборки.

#### Подбираем параметры

```python
alpha_arr = np.logspace(-3, 2, 21)
test_err = []
train_err = []
for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = (100,), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 10)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.semilogx(alpha_arr, train_err, 'b-o', label = 'train')
plt.semilogx(alpha_arr, test_err, 'r-o', label = 'test')
plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
pass
```

С уменьшением коэффициента регуляризатора ошибка уменьшается. Скорее всего, при большом значении альфа модель сглаживается и недообучается.

Минимальное значение ошибки:

```python
alpha_opt = alpha_arr[test_err == np.min(test_err)]
print(alpha_opt)
```

#### Классификатор при найденном оптимальном значении alpha:

```python
mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (100,),
                          solver = 'lbfgs', activation = 'logistic', random_state = 10)
mlp_model.fit(X_train, y_train)

print(np.min(train_err), np.min(test_err))
```

#### Подберём число нейронов:

```python
hidden_layer_sizes_arr = np.arange(50,300,50)
hidden_layer_sizes_arr
```

```python
test_err_hidden_layer = []
train_err_hidden_layer = []
for hidden_layer_sizes in hidden_layer_sizes_arr:
    mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = hidden_layer_sizes, 
                              solver = 'lbfgs', activation = 'logistic', random_state = 10)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err_hidden_layer.append(np.mean(y_train != y_train_pred))
    test_err_hidden_layer.append(np.mean(y_test != y_test_pred))
```

```python
plt.semilogx(hidden_layer_sizes_arr, train_err_hidden_layer, 'b-o', label = 'test')
plt.semilogx(hidden_layer_sizes_arr, test_err_hidden_layer, 'r-o', label = 'train')
plt.xlim([np.max(hidden_layer_sizes_arr), np.min(hidden_layer_sizes_arr)])
plt.title('Error vs. hidden_layer_sizes')
plt.xlabel('hidden_layer_sizes')
plt.ylabel('error')
plt.legend()
pass
```

#### Ошибка при оптимальном количестве нейронов

```python
print(np.min(train_err_hidden_layer), np.min(test_err_hidden_layer))
```

```python
hidden_layer = hidden_layer_sizes_arr[test_err_hidden_layer == np.min(test_err_hidden_layer)]
print(hidden_layer)
```

#### Сравним однослойную и двуслойную нейроную сеть при лучших показатлях

```python
mlp_model = MLPClassifier(hidden_layer_sizes = (int(hidden_layer), ), solver = 'lbfgs', 
                          activation = 'logistic', random_state = 10, alpha=alpha_opt)
mlp_model.fit(X_train, y_train)
print("Result of One-layer NN:")
print(mlp_model.score(X_train, y_train), mlp_model.score(X_test, y_test))
mlp_model_2l = MLPClassifier(hidden_layer_sizes = (int(hidden_layer), int(hidden_layer)), solver = 'lbfgs', 
                          activation = 'logistic', random_state = 10, alpha=alpha_opt)
mlp_model_2l.fit(X_train, y_train)
print("Result of Two-layer NN:")
print(mlp_model_2l.score(X_train, y_train), mlp_model_2l.score(X_test, y_test))
```

Результат улучшился на тренировочной выборке и ухудшился на тестовой, следовательно модель переобучается.

```python
range_ = np.arange(1,50,1)
range_
```

```python
test_err = []
train_err = []
oob_err = []
for r in range_:
    mlp_model = ensemble.RandomForestClassifier(n_estimators=r, random_state=41, n_jobs=-1,oob_score=True)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
    oob_err.append(1 - mlp_model.oob_score_)
```

```python
plt.plot(range_, train_err, 'b', label = 'train')
plt.plot(range_, test_err, 'r', label = 'test')
plt.plot(range_, oob_err, 'g', label = 'oob')
plt.title('Error vs. n_estimators')
plt.xlabel('n_estimators')
plt.ylabel('error')
plt.legend()
pass
```

### PCA for Extremely Randomized Trees

```python
import seaborn
from sklearn import ensemble
from sklearn.decomposition import PCA
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

Best 'n_estimators' for Test:  [13]

```python
n_estimators = 13
```

```python
range_ = np.arange(5,X_train.shape[1],5)
range_
```

```python
from tqdm import tqdm_notebook
```

```python
test_err = []
train_err = []

for r in tqdm_notebook(range_):
    pca = PCA(n_components=r)
    X_newTrain = pca.fit_transform(X_train)
    X_newTest = pca.transform(X_test)
    
    mlp_model = ensemble.ExtraTreesClassifier(n_estimators = n_estimators, n_jobs=-1)
    mlp_model.fit(X_newTrain, y_train)

    y_train_pred = mlp_model.predict(X_newTrain)
    y_test_pred = mlp_model.predict(X_newTest)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.plot(range_, train_err, 'b', label = 'train')
plt.plot(range_, test_err, 'r', label = 'test')
plt.title('Error vs. n_estimators')
plt.xlabel('PCA')
plt.ylabel('error')
plt.legend()
pass
```
